## Stacking Array

```python
array1 = np.array([[1,2],[3,4]])
array1
```

```python
array2 = np.array([[-1,-2],[-3,-4]])
array2
```

```python
#vertical stack

array3 = np.vstack((array1, array2))
array3
```

```python
#horizontal stack

array4 = np.hstack((array1, array2))
array4
```
